## Within subject correlation 

```python
withinSub_z = pd.DataFrame()
null = True
nIterations = 1000
i = 0
load = False

if load:
    withinSub_z = pd.read_csv('withinSub_z.csv', index_col=0)
else:
    for idx, condition in enumerate(condition_names):
        folder_path = main_path + condition
        ci_path = folder_path + '/cis/'
        for subject in subject_names:
            part1_path = [file for file in os.listdir(ci_path) if condition in file and 'part1' in file and subject in file][0]
            part2_path = [file for file in os.listdir(ci_path) if condition in file and 'part2' in file and subject in file][0]

            part1 = imageio.imread(ci_path + part1_path, as_gray=True)
            part2 = imageio.imread(ci_path + part2_path, as_gray=True)


            z = utils.z_transform(utils.ncc(part1, part2))
            withinSub_z.loc[i, 'SUBJECT'] = subject
            withinSub_z.loc[i, 'CONDITION'] = condition
            withinSub_z.loc[i, 'WITHINSUB_Z'] = z

            if null:
                ci_null_path = folder_path + '/cis_null_part2/'
                for iteration in range(nIterations):
                    ind_ci_path_null2 = [file for file in os.listdir(ci_null_path) if subject in file and 'iteration' + str(iteration) in file][0]
                    ci_image_null2 = imageio.imread(ci_null_path + ind_ci_path_null2, as_gray=True)
                    z_null = utils.z_transform(utils.ncc(part1, ci_image_null2))

                    withinSub_z.loc[i, str(iteration)] = z_null


            i+=1
    withinSub_z.to_csv('withinSub_z.csv')
```

```python
%%R -i condition_names -i main_path -i generate_nulls
if (generate_nulls) {
    devtools::install_github("rdotsch/rcicr", ref="development")

    library(rcicr)
    ct <- 1
    response_files <- c('wnull_U0013_inverted_allsubj_full.csv', 'wnull_U0013_upright_allsubj_full.csv', 'wnull_U0036_inverted_allsubj_full.csv',
                              'wnull_U0036_upright_allsubj_full.csv', 'wnull_U0074_inverted_allsubj_full.csv', 'wnull_U0074_upright_allsubj_full.csv',
                             'wnull_U0159_inverted_allsubj_full.csv', 'wnull_U0159_upright_allsubj_full.csv')

    base_images <- c('U0013_15opac_inv.png', 'U0013_15opac.png', 'U0036_15opac_inv.png', 'U0036_15opac.png',
                 'U0074_15opac_inv.png', 'U0074_15opac.png', 'U0159_15opac_inv.png', 'U0159_15opac.png')

    rdata_files <- c('U0013_15opac_inv_seed_1_time_Oct_07_2020_17_31.RData', 'U0013_15opac_seed_1_time_Oct_07_2020_17_29.RData',
                           'U0036_15opac_inv_seed_1_time_Oct_07_2020_17_34.RData', 'U0036_15opac_seed_1_time_Oct_07_2020_17_32.RData',
                           'U0074_15opac_inv_seed_1_time_Oct_07_2020_17_38.RData', 'U0074_15opac_seed_1_time_Oct_07_2020_17_36.RData',
                          'U0159_15opac_inv_seed_1_time_Oct_07_2020_17_41.RData', 'U0159_15opac_seed_1_time_Oct_07_2020_17_39.RData')

    for (condition_idx in 1:length(condition_names)){
        
        stimconditionpath <- paste(main_path, condition_names[condition_idx], sep='')
        print(condition_names[condition_idx])
        setwd(stimconditionpath)
        condition <- condition_names[condition_idx]
        rdata_path <- rdata_files[condition_idx]
        base_image <- base_images[condition_idx]
        response_file <- response_files[condition_idx]

        dat <- read.csv(response_file, sep = ',', stringsAsFactors = FALSE)
        load(rdata_path)
        allSub <- unique(dat$subject)

        for (iSub in allSub) {
            print(iSub)

            stimuli <- dat[dat$subject == iSub, ]$stimulus
            responses <- dat[dat$subject == iSub, ]$response
            filename <- paste('_', condition_names[condition_idx], '_', toString(iSub), '_full', sep="")

            ci <- generateCI(stimuli = stimuli, responses = responses, baseimage = "base", rdata = rdata_path, save_as_png=T, filename=filename, targetpath='./cis')
            
            for (iteration in 0:99) {
                print(iteration)
                iteration_str = paste('iteration', toString(iteration), sep="")
                responses_null <- dat[dat$subject == iSub, iteration_str]
                responses_null

                filename_null <- paste('_', condition_names[condition_idx], '_', toString(iSub), '_full_', iteration_str, sep="")
                ci <- generateCI(stimuli = stimuli, responses = responses_null, baseimage = "base", rdata = rdata_path, save_as_png=T, filename=filename_null, targetpath='./cis_null')

                ct <- ct + 1

            }
        }

    }

}
```

```python
betweenSub_z = pd.DataFrame()
betweenSub_z_null = pd.DataFrame()

null = True
nIterations = 1000
beg = 0
all_nulls_subject = pd.DataFrame()
load = True
if load:
    betweenSub_z = pd.read_csv('betweenSub_z.csv', index_col=0)
    all_nulls_subject = pd.read_csv('betweenSub_z_subject.csv', index_col=0)
else:

    for idx, condition in enumerate(condition_names):
        print(condition)
        z_df = pd.DataFrame(index=subject_names, columns=subject_names)
        z_df_null = pd.DataFrame(index=subject_names, columns=subject_names)
        condition_path = main_path + condition + '/'
        ci_path_condition = condition_path +'cis/' 
        all_nulls_condition = pd.DataFrame()
        null_i = 0
        for sub1 in subject_names:
                ind_ci_path1 = [file for file in os.listdir(ci_path_condition) if sub1 in file and 'part2' in file and not file.startswith('._')][0]
                print(ind_ci_path1)
                ci_image1 = imageio.imread(ci_path_condition + ind_ci_path1, as_gray=True)

                for sub2 in subject_names:

                    ind_ci_path2 = [file for file in os.listdir(ci_path_condition) if sub2 in file and 'part2' in file and not file.startswith('._')][0]
                    ci_image2 = imageio.imread(ci_path_condition + ind_ci_path2, as_gray=True)

                    if sub1 == sub2:
                        z_df.loc[sub1, sub2] = np.nan
                    else:
                        z = utils.z_transform(utils.ncc(ci_image1, ci_image2))
                        z_df.loc[sub1, sub2] = z

                        if null:
                            ci_null_path = condition_path + 'cis_null_part2/'
                            z_null_iterations = list()
                            for iteration in range(nIterations):
        #                         ind_ci_path_null2 = ci_null_path + 'ci__' + sub2 + '_iteration' + str(iteration) + '_part2.png'
                                ind_ci_path_null2 = [file for file in os.listdir(ci_null_path) if sub2 in file and 'iteration' + str(iteration) in file][0]
                                ci_image_null2 = imageio.imread(ci_null_path + ind_ci_path_null2, as_gray=True)

                                z_null = utils.z_transform(utils.ncc(ci_image1, ci_image_null2))
                                z_null_iterations.append(z_null)
                                all_nulls_subject.loc[beg, 'CONDITION'] = condition
                                all_nulls_subject.loc[beg, 'SUB1'] = sub1
                                all_nulls_subject.loc[beg, 'SUB2'] = sub2
                                all_nulls_subject.loc[beg, str(iteration)] = z_null

                            beg += 1

                            # sort iterations first
                            z_null_iterations_sorted = sorted(z_null_iterations)
    #                         all_nulls_condition = pd.DataFrame(z_null_iterations).T
                            all_nulls_condition = pd.concat([all_nulls_condition, pd.DataFrame(z_null_iterations).T])

                            # choose the 97 upper bound
                            upper_bound_znull = z_null_iterations_sorted[int(0.97 * nIterations)]
                            z_df_null.loc[sub1, sub2] = upper_bound_znull
                            null_i += 1

        mean_nulls = all_nulls_condition.loc[:, '0':str(nIterations-1)].mean(axis=0)
        mean_nulls_df = pd.DataFrame(mean_nulls).T

        betweenSub_z_null = pd.concat([betweenSub_z_null, mean_nulls_df], axis=0, ignore_index=True)
        z_df.to_csv(condition_path + condition + '_pairwiseZ' + '.csv')
        z_df_null.to_csv(condition_path + condition + '_pairwiseZnull'+ '.csv')

        lower_triangle = z_df.where(np.tril(np.ones(z_df.shape)).astype(np.bool)).replace(2.64,np.nan)
        betweenSub_z.loc[idx, 'CONDITION'] = condition
        betweenSub_z.loc[idx, 'BTW_MEAN_Z'] = np.mean(np.mean(lower_triangle))


    betweenSub_z = pd.concat([betweenSub_z, betweenSub_z_null], axis=1)
    betweenSub_z.to_csv('betweenSub_z.csv')
    all_nulls_subject.to_csv('betweenSub_z_subject.csv')
```

```python
betweenSub_z_masked = pd.DataFrame()
betweenSub_z_masked_null = pd.DataFrame()

null = True
nIterations = 1000
beg = 0
all_nulls_subject = pd.DataFrame()

load = True
if load:
    betweenSub_z_masked = pd.read_csv('betweenSub_z_masked.csv', index_col=0)
else:
    for idx, condition in enumerate(upright_names):
        print(condition)
        z_df = pd.DataFrame(index=subject_names, columns=subject_names)
        z_df_null = pd.DataFrame(index=subject_names, columns=subject_names)
        condition_path = main_path + condition + '/'
        ci_path_condition = condition_path +'cis/' 
        all_nulls_condition = pd.DataFrame()
        null_i = 0
        for sub1 in subject_names:
            ind_ci_path1 = [file for file in os.listdir(ci_path_condition) if sub1 in file and 'part2' in file and not file.startswith('._')][0]
            print(ind_ci_path1)
            ci_image1 = imageio.imread(ci_path_condition + ind_ci_path1, as_gray=True)
            ci_image1 = ci_image1[black_coords[condition]]
            for sub2 in subject_names:

                ind_ci_path2 = [file for file in os.listdir(ci_path_condition) if sub2 in file and 'part2' in file and not file.startswith('._')][0]
                ci_image2 = imageio.imread(ci_path_condition + ind_ci_path2, as_gray=True)
                ci_image2 = ci_image2[black_coords[condition]]
                if sub1 == sub2:
                    z_df.loc[sub1, sub2] = np.nan
                else:
                    z = utils.z_transform(utils.ncc(ci_image1, ci_image2))
                    z_df.loc[sub1, sub2] = z

                    if null:
                        ci_null_path = condition_path + 'cis_null_part2/'
                        z_null_iterations = list()
                        for iteration in range(nIterations):
            #                         ind_ci_path_null2 = ci_null_path + 'ci__' + sub2 + '_iteration' + str(iteration) + '_part2.png'
                            ind_ci_path_null2 = [file for file in os.listdir(ci_null_path) if sub2 in file and 'iteration' + str(iteration) in file][0]
                            ci_image_null2 = imageio.imread(ci_null_path + ind_ci_path_null2, as_gray=True)
                            ci_image_null2 = ci_image_null2[black_coords[condition]]


                            z_null = utils.z_transform(utils.ncc(ci_image1, ci_image_null2))
                            z_null_iterations.append(z_null)
                            all_nulls_subject.loc[beg, 'CONDITION'] = condition
                            all_nulls_subject.loc[beg, 'SUB1'] = sub1
                            all_nulls_subject.loc[beg, 'SUB2'] = sub2
                            all_nulls_subject.loc[beg, str(iteration)] = z_null

                        beg += 1

                        # sort iterations first
                        z_null_iterations_sorted = sorted(z_null_iterations)
            #                 print(z_null_iterations)
                        all_nulls_condition = pd.concat([all_nulls_condition, pd.DataFrame(z_null_iterations).T])
                        # choose the 97 upper bound
                        upper_bound_znull = z_null_iterations_sorted[int(0.97 * nIterations)]
                        z_df_null.loc[sub1, sub2] = upper_bound_znull
                        null_i += 1
        mean_nulls = all_nulls_condition.loc[:, '0':str(nIterations-1)].mean(axis=0)
        mean_nulls_df = pd.DataFrame(mean_nulls).T

        betweenSub_z_masked_null = pd.concat([betweenSub_z_masked_null, mean_nulls_df], axis=0, ignore_index=True)
    #     z_df.to_csv(condition_path + condition + '_pairwiseZ' + '.csv')
    #     z_df_null.to_csv(condition_path + condition + '_pairwiseZnull'+ '.csv')

        lower_triangle = z_df.where(np.tril(np.ones(z_df.shape)).astype(np.bool)).replace(2.64,np.nan)
        betweenSub_z_masked.loc[idx, 'CONDITION'] = condition
        betweenSub_z_masked.loc[idx, 'BTW_MEAN_Z'] = np.mean(np.mean(lower_triangle))


    betweenSub_z_masked = pd.concat([betweenSub_z_masked, betweenSub_z_masked_null], axis=1)
    betweenSub_z_masked.to_csv('betweenSub_z_masked.csv')
    
```

```python
def calc_p(null, empirical):
    p = np.sum(null < empirical)/len(null)
    if p < .5:
        twoTailedP = p * 2;
    else:
        twoTailedP = (1 - p) * 2;
    return twoTailedP
```
